## Objective
__To minimiza the cost function for J(w, b) = wx + b where b is zero so J(w) = wx__

## Data
For this example let's take 50 total data points

```python
import numpy as np
import matplotlib.pyplot as plt
# let's generate 50 datapoints from 1 to 50
x = np.arange(1, 51)
y = np.arange(1, 51)

# Plot the data points
plt.plot(x, y, marker = 'x', color = 'red')
plt.title("Dataset plot")
plt.xlabel('x')
plt.ylabel('y')
```

## develop the model
Our cost function is ___J(x) = wx___ so let's try different value of w and minimize __J(w)__

```python
# Generate 100 trial values of w
w = np.linspace(-5, 7, num = 100)
```

```python
x - y
```

```python
def f(x, y, w):
    j = np.zeros(100)
    for i, w_i in np.ndenumerate(w):
       j[i] = np.sum((w_i * x - y)**2)/(2 *len(x))
    return j
```

```python
# determine error for each value of w
error = f(x, y, w)
plt.plot(w, error)
plt.title('Cost Function')
plt.xlabel('w')
plt.ylabel('J(w)')
```
